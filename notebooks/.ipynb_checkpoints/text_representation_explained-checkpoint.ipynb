# 📖 Introduction to Text Representation

Welcome to this notebook on Text Representation! 

Machine Learning models (like Neural Networks, Logistic Regression, etc.) cannot process raw text. They only understand **numbers** and **math**. Therefore, before we can train an AI on text data, we must convert our sentences into numerical formats (usually vectors or matrices).

In this notebook, we will explore the foundational techniques for text representation:
1. **Tokenization**
2. **Vocabulary Creation**
3. **One-Hot Encoding (Manual & Automated)**
4. **Bag of Words (BoW)**

## 1. Defining the Corpus
A **corpus** (plural: corpora) is simply a collection of text documents. In the real world, this could be a dataset of 10,000 movie reviews or 1 million tweets. Here, our corpus consists of three simple sentences.

# Original Corpus

```python
corpus = [
    "machine learning transform artificial intelligence",
    "machine learning use neural network",
    "deep learning subset machine learning"
]
```

## 2. Tokenization
**Tokenization** is the process of breaking down a large block of text into smaller, manageable pieces called **tokens**. Most commonly, tokens are individual words.

Here, we are using Python's built-in `.split()` method to split the sentences by spaces. *(Note: Professional tokenizers also handle punctuation, special characters, and lowercasing!)*

# Tokenization

```python
tokenized = []

for doc in corpus:
    words = doc.split()
    tokenized.append(words)

print(tokenized)
```

## 3. Creating the Vocabulary
To represent our text mathematically, we need to know every unique word that exists across our entire corpus. This collection of unique words is our **Vocabulary**. 

We use a Python `set()` here because sets automatically remove duplicate words.

# Create Vocabulary Manually

```python
vocabulary = set()

for document in tokenized:
    for word in document:
        vocabulary.add(word)

print(vocabulary)
```

Sorting the vocabulary ensures that our final vector representations are consistent every time we run the code. Alphabetical order is standard.

# Sort Vocabulary

```python
vocabulary = sorted(vocabulary)

print(vocabulary)
```

Now, we assign a unique numerical index to each word in our sorted dictionary. The length of this vocabulary determines the length of our vectors.

# Assign Index to Each Word

```python
for i, word in enumerate(vocabulary):
    print(i+1, "->", word)
```

## 4. One-Hot Encoding
**One-Hot Encoding** creates a binary vector for each document. 
* If a vocabulary word is present in the document, it gets a `1`.
* If it is missing, it gets a `0`.

Let's create the vector for our first document (`D1`).

# Create One-Hot Vector for D1

```python
doc = tokenized[0]

vector = []

for word in vocabulary:
    if word in doc:
        vector.append(1)
    else:
        vector.append(0)

print("Tokens:", doc)
print("Vector:", vector)
```

```python
doc = tokenized[1]

vector = []

for word in vocabulary:
    if word in doc:
        vector.append(1)
    else:
        vector.append(0)

print("Tokens:", doc)
print("Vector:", vector)
```

```python
doc = tokenized[2]

vector = []

for word in vocabulary:
    if word in doc:
        vector.append(1)
    else:
        vector.append(0)

print("Tokens:", doc)
print("Vector:", vector)
```

Instead of doing this document by document, we can write a nested loop to create a **Matrix** (a 2D list) containing all the vectors for our entire corpus.

# Create Complete One-Hot Matrix

```python
one_hot_matrix = []

for doc in tokenized:
    vector = []
    for word in vocabulary:
        if word in doc:
            vector.append(1)
        else:
            vector.append(0)
    one_hot_matrix.append(vector)

for row in one_hot_matrix:
    print(row)
```

## 5. Automated Encoding with Scikit-Learn
Writing manual loops for tokenization and encoding is great for learning, but it is too slow for large datasets. 

Python's `scikit-learn` library provides highly optimized tools to do this in just two lines of code! The `MultiLabelBinarizer` takes our tokenized lists and instantly converts them into a One-Hot Matrix.

# MultiLabelBinarizer

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

one_hot = mlb.fit_transform(tokenized)

print("Vocabulary Classes:", mlb.classes_)
print("\nMatrix:\n", one_hot)
```

## 6. Bag of Words (BoW)
One-Hot encoding has a flaw: it only tells us *if* a word exists (0 or 1). It doesn't capture the frequency of the word. If the word "machine" appears 10 times in a document, it's probably very important, but One-Hot encoding still just gives it a `1`.

**Bag of Words (BoW)** solves this by counting the frequency of each word. Let's try it on a new, slightly larger corpus.

# Bag of words

```python
corpus = [
    "machine learning artificial intelligence machine learning",
    "deep learning neural network machine learning",
    "artificial intelligence use machine learning solve problem",
    "neural network deep learning require large dataset",
    "machine learning artificial intelligence transform healthcare education"
]
```

```python
tokenized = [doc.split() for doc in corpus]
tokenized
```

```python
vocabulary = sorted(
    set(word
        for doc in tokenized
        for word in doc)
)
vocabulary
```

The code below generates a **Binary Bag of Words**, which is structurally identical to the One-Hot representation we did earlier.

```python
binary_bow = []

for doc in tokenized:
    vector = []
    for word in vocabulary:
        if word in doc:
            vector.append(1)
        else:
            vector.append(0)
    binary_bow.append(vector)

for row in binary_bow:
    print(row)
```

### CountVectorizer (The Industry Standard)
Instead of manual loops, we use `CountVectorizer` from `scikit-learn`. 

Notice the output below: The numbers represent **Term Frequency**. In the first document, "machine" and "learning" both appear twice, so their value in the vector is `2`.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Initializes the vectorizer (by default, it counts frequency)
vectorizer = CountVectorizer()

# fit_transform does two things: creates the vocabulary AND converts text to vectors
X = vectorizer.fit_transform(corpus)

print("Vocabulary:", vectorizer.get_feature_names_out())
print("\nFrequency Bag of Words Matrix:\n")
print(X.toarray())
```

If we pass the argument `binary=True` to `CountVectorizer`, it stops counting frequency and just returns `1` (present) or `0` (absent), giving us the exact same output as our manual Binary Bag of Words / One-Hot logic!

```python
from sklearn.feature_extraction.text import CountVectorizer

# Forces the vectorizer to only return 0s and 1s
vectorizer = CountVectorizer(binary=True)

X = vectorizer.fit_transform(corpus)

print("Binary Bag of Words Matrix:\n")
print(X.toarray())
```
