# Original Corpus

```python
corpus = [
    "machine learning transform artificial intelligence",
    "machine learning use neural network",
    "deep learning subset machine learning"
]
```

# Tokenization

```python
tokenized = []

for doc in corpus:
    words = doc.split()
    tokenized.append(words)

print(tokenized)
```

# Create Vocabulary Manually

```python
vocabulary = set()

for document in tokenized:
    for word in document:
        vocabulary.add(word)

print(vocabulary)
```

# Sort Vocabulary

```python
vocabulary = sorted(vocabulary)

print(vocabulary)
```

# Assign Index to Each Word

```python
for i, word in enumerate(vocabulary):
    print(i+1, "->", word)
```

# Create One-Hot Vector for D1

```python
doc = tokenized[0]

vector = []

for word in vocabulary:

    if word in doc:
        vector.append(1)
    else:
        vector.append(0)

print(vector)
```

```python
doc = tokenized[1]

vector = []

for word in vocabulary:

    if word in doc:
        vector.append(1)
    else:
        vector.append(0)

print(vector)
```

```python
doc = tokenized[2]

vector = []

for word in vocabulary:

    if word in doc:
        vector.append(1)
    else:
        vector.append(0)

print(vector)
```

# Create Complete One-Hot Matrix

```python
one_hot_matrix = []

for doc in tokenized:

    vector = []

    for word in vocabulary:

        if word in doc:
            vector.append(1)
        else:
            vector.append(0)

    one_hot_matrix.append(vector)

print(one_hot_matrix)
```

# MultiLabelBinarizer

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()

one_hot = mlb.fit_transform(tokenized)

print(mlb.classes_)
print(one_hot)
```

# Bag of words

```python
corpus = [
    "machine learning artificial intelligence machine learning",

    "deep learning neural network machine learning",

    "artificial intelligence use machine learning solve problem",

    "neural network deep learning require large dataset",

    "machine learning artificial intelligence transform healthcare education"
]
```

```python
tokenized = [doc.split() for doc in corpus]
tokenized
```

```python
vocabulary = sorted(
    set(word
        for doc in tokenized
        for word in doc)
)
vocabulary
```

```python
binary_bow = []

for doc in tokenized:

    vector = []

    for word in vocabulary:

        if word in doc:
            vector.append(1)
        else:
            vector.append(0)

    binary_bow.append(vector)

for row in binary_bow:
    print(row)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(corpus)

print(X.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(binary=True)

X = vectorizer.fit_transform(corpus)

print(X.toarray())
```
